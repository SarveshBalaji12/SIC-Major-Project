# Feature Engineering

We'll implement a comprehensive feature engineering pipeline including:

1. **Textual Features**
   - TF-IDF Vectorization
   - Word Embeddings (IndicBERT/mBERT)

2. **Structural Features**
   - Headline length metrics
   - Word count analysis
   - Statistical measures

3. **Semantic Features**
   - Contextual embeddings
   - Language-specific features

4. **Dimensionality Reduction**
   - PCA visualization
   - t-SNE analysis

```python
# 1. Structural Features

# Word count function
def get_word_count(text):
    return len(str(text).split())

# Add structural features
data_with_districts_removed['word_count'] = data_with_districts_removed['headline'].apply(get_word_count)
data_with_districts_removed['avg_word_length'] = data_with_districts_removed['headline'].apply(lambda x: np.mean([len(word) for word in str(x).split()]))

# Visualize structural features
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Word count distribution
sns.histplot(data=data_with_districts_removed, x='word_count', bins=30, ax=axes[0,0])
axes[0,0].set_title('Distribution of Word Count')
axes[0,0].set_xlabel('Word Count')

# Average word length distribution
sns.histplot(data=data_with_districts_removed, x='avg_word_length', bins=30, ax=axes[0,1])
axes[0,1].set_title('Distribution of Average Word Length')
axes[0,1].set_xlabel('Average Word Length')

# Word count by category
sns.boxplot(data=data_with_districts_removed, x='category', y='word_count', ax=axes[1,0])
axes[1,0].set_xticklabels(axes[1,0].get_xticklabels(), rotation=45, ha='right')
axes[1,0].set_title('Word Count by Category')

# Avg word length by category
sns.boxplot(data=data_with_districts_removed, x='category', y='avg_word_length', ax=axes[1,1])
axes[1,1].set_xticklabels(axes[1,1].get_xticklabels(), rotation=45, ha='right')
axes[1,1].set_title('Average Word Length by Category')

plt.tight_layout()
plt.show()

# Print summary statistics
print("\nStructural Features Summary:")
print("\nWord Count Statistics:")
print(data_with_districts_removed['word_count'].describe())
print("\nAverage Word Length Statistics:")
print(data_with_districts_removed['avg_word_length'].describe())
```

```python
# 2. Textual Features - TF-IDF

from sklearn.feature_extraction.text import TfidfVectorizer

# Create TF-IDF features
tfidf = TfidfVectorizer(max_features=1000, stop_words=None)  # No stop words as we need to handle Kannada
tfidf_features = tfidf.fit_transform(data_with_districts_removed['headline'])

# Get top terms for each category
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(data_with_districts_removed['category'])

def print_top_terms(vectorizer, feature_matrix, categories, n_top=10):
    for i, category in enumerate(le.classes_):
        top_indices = np.argsort(np.asarray(feature_matrix[y == i].mean(axis=0)).ravel())[-n_top:]
        top_terms = [vectorizer.get_feature_names_out()[j] for j in top_indices]
        print(f"\nTop {n_top} terms for category '{category}':")
        print(", ".join(top_terms))

print("\nTop TF-IDF terms by category:")
print_top_terms(tfidf, tfidf_features, le.classes_)

# Dimensionality Reduction for Visualization
from sklearn.decomposition import PCA

# Convert sparse matrix to dense for PCA
pca = PCA(n_components=2)
tfidf_2d = pca.fit_transform(tfidf_features.toarray())

# Plot PCA visualization
plt.figure(figsize=(12, 8))
scatter = plt.scatter(tfidf_2d[:, 0], tfidf_2d[:, 1], c=y, cmap='tab10', alpha=0.6)
plt.title('PCA of TF-IDF Features')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.tight_layout()
plt.show()

# Print explained variance ratio
print("\nExplained variance ratio:", pca.explained_variance_ratio_)
```

```python
# 3. Dimensionality Reduction - t-SNE

from sklearn.manifold import TSNE

# Apply t-SNE
tsne = TSNE(n_components=2, random_state=42)
tsne_features = tsne.fit_transform(tfidf_features.toarray())

# Plot t-SNE visualization
plt.figure(figsize=(12, 8))
scatter = plt.scatter(tsne_features[:, 0], tsne_features[:, 1], c=y, cmap='tab10', alpha=0.6)
plt.title('t-SNE Visualization of TF-IDF Features')
plt.xlabel('First t-SNE Component')
plt.ylabel('Second t-SNE Component')
plt.legend(handles=scatter.legend_elements()[0], labels=le.classes_, title="Categories", bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```
